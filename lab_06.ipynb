# Task 1 – Hill Climbing
Use the Hill Climbing algorithm to find the maximum value of the function <br>
f(x) = -x2 + 6x <br>
Range: 0 ≤ x ≤ 6 <br>
**Instructions:** <br>
• Start from a random value of x. <br>
• Check the neighbor values (x+1 and x−1). <br>
• Move to the neighbor with the higher value of f(x). <br>
• Stop when no better neighbor exists. <br>
**Output should show:** <br>
• Initial value of x <br>
• Value of f(x) at each step <br>
• Final optimal value.

```python
import random

def function(x):
    return -(x**2) + 6*x

def hill_climb():
    x = random.randint(0,6)
    print("Starting point:", x)

    while True:
        current_val = function(x)
        print("x =", x, " f(x) =", current_val)

        left = x - 1
        right = x + 1

        best_x = x
        best_val = current_val

        if left >= 0:
            if function(left) > best_val:
                best_val = function(left)
                best_x = left

        if right <= 6:
            if function(right) > best_val:
                best_val = function(right)
                best_x = right

        if best_x == x:
            break

        x = best_x

    print("\nBest solution found:")
    print("x =", x)
    print("Maximum f(x) =", function(x))


hill_climb()
```

# Task 2 – Beam Search
Use Beam Search to reach the number 20 starting from 1. <br>
**Allowed operations:** <br>
• x → x + 2 <br>
• x → x + 3 <br>
• x → x × 2 <br>
**Heuristic function:** <br>
h(n) = |20 − current number| <br>
Beam width: k = 2 <br>
**Instructions:** <br>
• Start from number 1. <br>
• Generate the next possible numbers using the allowed operations. <br>
• At each level keep only the best 2 states with the lowest heuristic value. <br>
• Continue until the goal number 20 is reached. <br>
Output should show: <br>
• Numbers explored at each level <br>
• Final path to reach 20.

```python
goal = 20
beam_width = 2

def h(n):
    return abs(20 - n)

def expand(n):
    return [n+2, n+3, n*2]

def beam_search(start):
    beam = [[start]]
    level = 0
    while beam:
        print("Level", level, ":", [p[-1] for p in beam])
        next_states = []
        for path in beam:
            node = path[-1]
            if node == goal:
                return path
            children = expand(node)
            for c in children:
                if c <= 40:
                    new_path = path + [c]
                    next_states.append(new_path)
        next_states.sort(key=lambda p: h(p[-1]))
        beam = next_states[:beam_width]
        level += 1
    return None

result = beam_search(1)
if result:
    print("\nPath to reach 20:")
    print(" -> ".join(str(x) for x in result))
else:
    print("Goal not found")
```

# Task 3 – Genetic Algorithm
Use a Genetic Algorithm to maximize the function <br>
f(x) = x2 + 2x <br>
Range: 0 ≤ x ≤ 31 <br>
**Instructions:** <br>
• Use binary representation for chromosomes. <br>
• Generate a random initial population. <br>
• Calculate fitness using the function f(x). <br>
• Apply selection, crossover, and mutation. <br>
• Run the algorithm for 15 generations. <br>
Output should show: <br>
• Best chromosome <br>
• Best value of x <br>
• Best fitness value. <br>

```python
import random

def decode(ch):
    return int(ch, 2)

def fitness(ch):
    x = decode(ch)
    return x*x + 2*x

def random_chromosome():
    return "".join(random.choice("01") for _ in range(5))

def select(pop):
    a = random.choice(pop)
    b = random.choice(pop)
    return a if fitness(a) > fitness(b) else b

def crossover(p1, p2):
    point = random.randint(1,4)
    c1 = p1[:point] + p2[point:]
    c2 = p2[:point] + p1[point:]
    return c1, c2

def mutate(ch, rate):
    new = ""
    for bit in ch:
        if random.random() < rate:
            new += "1" if bit == "0" else "0"
        else:
            new += bit
    return new


population_size = 6
generations = 15
mutation_rate = 0.02
population = [random_chromosome() for _ in range(population_size)]
for g in range(generations):
    new_pop = []
    while len(new_pop) < population_size:
        parent1 = select(population)
        parent2 = select(population)
        child1, child2 = crossover(parent1, parent2)
        child1 = mutate(child1, mutation_rate)
        child2 = mutate(child2, mutation_rate)
        new_pop.append(child1)
        new_pop.append(child2)
    population = new_pop[:population_size]


best = max(population, key=fitness)
print("Best chromosome:", best)
print("Best x:", decode(best))
print("Best fitness:", fitness(best))
```
