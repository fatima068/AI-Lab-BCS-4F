#Task 1: Minimax Algorithm Implementation
1. Implement the Minimax algorithm for the tree above.
2. Compute and print:
  - Minimax values for all nodes
  - Order in which nodes are visited
3. Add depth-limiting (depth = 2) and show updated Minimax values.

```python
import math

class Minimax:
    def __init__(self):
        self.leaves = [4, 7, 2, 5, 1, 8, 3, 6]
        self.visited_order = []

    def solve(self, depth, node_index, is_max_player, path_name):
        self.visited_order.append(path_name)
        if depth == 3:
            return self.leaves[node_index]

        if is_max_player:
            best_val = -math.inf
            for i in range(2):
                child_name = f"L{depth+1}_N{node_index * 2 + i}"
                val = self.solve(depth + 1, node_index * 2 + i, False, child_name)
                best_val = max(best_val, val)
            return best_val
        else:
            best_val = math.inf
            for i in range(2):
                child_name = f"L{depth+1}_N{node_index * 2 + i}"
                val = self.solve(depth + 1, node_index * 2 + i, True, child_name)
                best_val = min(best_val, val)
            return best_val

game = Minimax()
node_vals = {}
root = game.solve(0, 0, True, "Root")
node_vals["Root"] = root

node_vals["L1_N0"] = game.solve(1, 0, False, "L1_N0")
node_vals["L1_N1"] = game.solve(1, 1, False, "L1_N1")

node_vals["L2_N0"] = game.solve(2, 0, True, "L2_N0")
node_vals["L2_N1"] = game.solve(2, 1, True, "L2_N1")
node_vals["L2_N2"] = game.solve(2, 2, True, "L2_N2")
node_vals["L2_N3"] = game.solve(2, 3, True, "L2_N3")

for i in range(8):
    node_vals[f"L3_N{i}"] = game.leaves[i]

print("Minimax values (all nodes):")
for k in node_vals:
    print(k, "=", node_vals[k])

print("\nOrder of nodes visited:")
for i in range(len(game.visited_order)):
    if i == len(game.visited_order) - 1:
        print(game.visited_order[i])
    else:
        print(game.visited_order[i], end=" -> ")

print("\ndepth = 2")
limited_vals = {}

limited_vals["L2_N0"] = max(game.leaves[0], game.leaves[1])
limited_vals["L2_N1"] = max(game.leaves[2], game.leaves[3])
limited_vals["L2_N2"] = max(game.leaves[4], game.leaves[5])
limited_vals["L2_N3"] = max(game.leaves[6], game.leaves[7])

limited_vals["L1_N0"] = min(limited_vals["L2_N0"], limited_vals["L2_N1"])
limited_vals["L1_N1"] = min(limited_vals["L2_N2"], limited_vals["L2_N3"])

limited_vals["Root"] = max(limited_vals["L1_N0"], limited_vals["L1_N1"])

print("Updated minimax values (depth = 2):")
for k in limited_vals:
    print(k, "=", limited_vals[k])
```

#Task 2: Alpha-Beta Pruning
1. Implement Minimax with Alpha-Beta pruning on the same tree.
2. Track alpha and beta values.
3. Print:
- Minimax values for all nodes
- Nodes that are pruned
4. Compare the number of nodes visited with standard Minimax and explain why
pruning reduces computation.

```python
import math
class AlphaBeta:
    def __init__(self):
        self.leaves = [4, 7, 2, 5, 1, 8, 3, 6]
        self.visited = []
        self.pruned = []
        self.node_vals = {}

    def solve(self, depth, node_index, is_max, alpha, beta, name):
        self.visited.append(name)
        if depth == 3:
            val = self.leaves[node_index]
            self.node_vals[name] = val
            return val

        if is_max:
            best = -math.inf
            for i in range(2):
                child = f"L{depth+1}_N{node_index*2+i}"
                val = self.solve(depth+1, node_index*2+i, False, alpha, beta, child)
                best = max(best, val)
                alpha = max(alpha, best)

                if beta <= alpha:
                    for j in range(i+1, 2):
                        self.pruned.append(f"L{depth+1}_N{node_index*2+j}")
                    break
            self.node_vals[name] = best
            return best
        else:
            best = math.inf
            for i in range(2):
                child = f"L{depth+1}_N{node_index*2+i}"
                val = self.solve(depth+1, node_index*2+i, True, alpha, beta, child)
                best = min(best, val)
                beta = min(beta, best)
                if beta <= alpha:
                    for j in range(i+1, 2):
                        self.pruned.append(f"L{depth+1}_N{node_index*2+j}")
                    break
            self.node_vals[name] = best
            return best

game = AlphaBeta()
ans = game.solve(0, 0, True, -math.inf, math.inf, "Root")
print("final minimax value:", ans)
print("\nall node values:")

for k in game.node_vals:
    print(k, "=", game.node_vals[k])

print("\nvisited nodes count:", len(game.visited))
print("visited order:")

for i in range(len(game.visited)):
    if i == len(game.visited) - 1:
        print(game.visited[i])
    else:
        print(game.visited[i], end=" -> ")

print("pruned nodes:")

if len(game.pruned) == 0:
    print("None")
else:
    for p in game.pruned:
        print(p)



# comparison part (task 4)

mini = Minimax()
mini.solve(0, 0, True, "Root")

print("\nComparison ")
print("Minimax visited:", len(mini.visited_order))
print("AlphaBeta visited:", len(game.visited))
if len(game.visited) < len(mini.visited_order):
    print("pruning helped reduce nodes")
else:
    print("No pruning here")
```

#Task 4 Basic CSP Modelling
Create a CSP model with three variables:
* Variables: A, B, C
* Domain: {0,1,2,3}

Apply the following constraints:
1. A ≠ B
2. B ≠ C
3. A + B ≤ 4

Requirements:
* Define variables using model.new_int_var()
* Add constraints using model.add()
* Solve the model
* Print the values of A, B, and C

```python
from ortools.sat.python import cp_model

model = cp_model.CpModel()
A = model.NewIntVar(0, 3, 'A')
B = model.NewIntVar(0, 3, 'B')
C = model.NewIntVar(0, 3, 'C')
model.Add(A != B)
model.Add(B != C)
model.Add(A + B <= 4)

solver = cp_model.CpSolver()
status = solver.Solve(model)
if status == cp_model.FEASIBLE or status == cp_model.OPTIMAL:
    print("Solution found:")
    print("A =", solver.Value(A))
    print("B =", solver.Value(B))
    print("C =", solver.Value(C))
else:
    print("No solution found")
```

#Task 5 Print All Possible Solutions
Modify the program from Task 1 to print all possible valid solutions.

Hint:
Use the solver parameter:
solver.parameters.enumerate_all_solutions = True

Also implement a solution callback class to print each solution.
Output should show all valid assignments of A, B, and C

```python
class SolutionPrinter(cp_model.CpSolverSolutionCallback):
    def __init__(self, variables):
        cp_model.CpSolverSolutionCallback.__init__(self)
        self.vars = variables
        self.solution_count = 0

    def OnSolutionCallback(self):
        self.solution_count += 1
        vals = [self.Value(v) for v in self.vars]
        print(f"Solution {self.solution_count}: A={vals[0]}, B={vals[1]}, C={vals[2]}")

model = cp_model.CpModel()
A = model.NewIntVar(0, 3, 'A')
B = model.NewIntVar(0, 3, 'B')
C = model.NewIntVar(0, 3, 'C')
model.Add(A != B)
model.Add(B != C)
model.Add(A + B <= 4)

solver = cp_model.CpSolver()
solver.parameters.enumerate_all_solutions = True
solution_printer = SolutionPrinter([A, B, C])
status = solver.Solve(model, solution_printer)

print(f"Total solutions found: {solution_printer.solution_count}")
```

#Task 6 Simple Optimization Problem
Create a CSP model with:

Variables: x, y, z

Domain: 0–20

Constraints:
* x + 2y + z ≤ 20
* 3x + y ≤ 18

Objective Function: maximize 4x + 2y + z

Requirements:
* Define variables
* Add constraints
* Define the objective function
* Solve the model

Print:
* Optimal value
* Values of x, y, and z

```python
model = cp_model.CpModel()
x = model.NewIntVar(0, 20, 'x')
y = model.NewIntVar(0, 20, 'y')
z = model.NewIntVar(0, 20, 'z')
model.Add(x + 2*y + z <= 20)
model.Add(3*x + y <= 18)
model.Maximize(4*x + 2*y + z)

solver = cp_model.CpSolver()
status = solver.Solve(model)
if status == cp_model.OPTIMAL or status == cp_model.FEASIBLE:
    print("Optimal value of objective:", solver.ObjectiveValue())
    print("x =", solver.Value(x))
    print("y =", solver.Value(y))
    print("z =", solver.Value(z))
else:
    print("No solution found")
```

#Task 7 N-Queens Variation
Write a Python program using OR-Tools to solve the 4-Queens problem.
Requirements:
1. No two queens share the same row
2. No two queens share the same diagonal
3. Print at least one valid solution

Example output format:

_ Q _ _

_ _ _ Q

Q _ _ _

_ _ Q _

```python
model = cp_model.CpModel()
N = 5
queens = [model.NewIntVar(0, N-1, f'q{i}') for i in range(N)]
model.AddAllDifferent(queens)

for i in range(N):
    for j in range(i+1, N):
        model.AddAbsEquality(model.NewIntVar(1, N-1, f'diff{i}{j}'), queens[i] - queens[j])
        model.Add(queens[i] - queens[j] != j - i)
        model.Add(queens[i] - queens[j] != i - j)

solver = cp_model.CpSolver()
status = solver.Solve(model)

if status == cp_model.FEASIBLE or status == cp_model.OPTIMAL:
    board = []
    for i in range(N):
        row = ["_"] * N
        row[solver.Value(queens[i])] = "Q"
        board.append(" ".join(row))
    print(" valid solution:\n")
    for row in board:
        print(row)
else:
    print("No solution found")
```
