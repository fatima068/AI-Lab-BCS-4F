**QUESTION 1**

```python
studentName = input("enter student name: ")
marks = int(input("enter marks: "))

print ("student name: ", studentName, "\nmarks: ", marks)

if marks >= 85:
    print("grade: A")
elif marks >= 70:
    print ("grade: B")
elif marks >= 50:
    print ("grade: C")
else:
    print("grade: Fail")
```

**QUESTION 2**

```python
count = 0
n = int(input("Enter a number: "))
print("even numbers: ")

for i in range(2, n+1):
    if i % 2 == 0:
        count += 1
        print(i, " ")

print("total even numbers: ", count)
```

**QUESTION 3**

```python
dict = {}

for i in range(3):
    name = input("Enter name: ")
    marks = int(input("Enter marks: "))
    dict[name] = marks

print("\nstudent records: ")
for i in dict:
    print(i, ": ", dict[i])
```

**QUESTION 4**

```python
choice = 0

while choice != 3:
  choice = int(input("1. add two numbers \n2. subtract two numbers \n3. exit \nChoice: "))
  if choice == 1:
    num1 = int(input("first number: "))
    num2 = int(input("second number: "))
    print("sum: ", num1 + num2)
  elif choice == 2:
    num1 = int(input("first number: "))
    num2 = int(input("second number: "))
    print("difference: ", num1 - num2)
  elif choice == 3:
    print("Goodbye")
  else:
    print("Invalid choice")
```

**QUESTION 5**

```python
n = int(input("how many numbers do u want to calcuate avg of? "))
marks = []

for i in range(n):
    userIn = int(input(f"enter number {i +1}: "))
    marks.append(userIn)

def calcAvg(marks):
    sumMarks = sum(marks)
    avg = sumMarks/n
    return avg

ans = calcAvg(marks)
print("average marks: ", ans)
```
