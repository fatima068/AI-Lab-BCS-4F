#TASK 1
Design a Vehicle Rental System using OOP.
Requirements:
1. Create a class Vehicle
2. Attributes: vehicle_id, brand, rent_per_day
3. Methods: display_details(), calculate_rent(days)
4. Create at least two vehicle objects
5. Calculate rent for different days

```python
class Vehicle:
  def __init__(self, vehicleID, brand, rentPerDay):
    self.__vehicleID = vehicleID
    self.__brand = brand
    self.__rentPerDay = rentPerDay

  def displayDetails(self):
    print(f"Vehicle ID: {self.__vehicleID}\nBrand: {self.__brand}\nRent per day: {self.__rentPerDay}")

  def calculateRent(self, days):
    rent = self.__rentPerDay* days
    print(f"rent for {days} days: {rent}")

v1 = Vehicle("v1", "mercedes", 100)
v2 = Vehicle("v2", "toyota", 50)
v1.displayDetails()
v1.calculateRent(5)
v2.displayDetails()
v2.calculateRent(3)
```

#TASK 2
Design an Employee Payroll System using inheritance.
Requirements:
1. Base class Employee (name, emp_id)
2. Child classes:
 - FullTimeEmployee (monthly_salary)
 - PartTimeEmployee (hours_worked, hourly_rate)
3. Override calculate_salary() method
4. Create objects and display salary

```python
class Employee:
  def __init__(self, name, empID):
    self._name = name
    self._empID = empID

  def calculateSalary(self):
    print("calculating salary for base employee")

class FullTimeEmployee(Employee):
  def __init__(self, name, empID, monthlySalary):
    super().__init__(name, empID)
    self.__monthlySalary = monthlySalary

  def calculateSalary(self):
    print(f"salary for full time employee {self._name}: {self.__monthlySalary}")

class PartTimeEmployee(Employee):
  def __init__(self, name, empID, hoursWorked, hourlyRate):
    super().__init__(name, empID)
    self.__hoursWorked = hoursWorked
    self.__hourlyRate = hourlyRate

  def calculateSalary(self):
    print(f"salary for part time employee {self._name}: {self.__hoursWorked*self.__hourlyRate}")

baseEmp = Employee("base", "e1")
baseEmp.calculateSalary()
fullEmp = FullTimeEmployee("full time", "e2", 10000)
fullEmp.calculateSalary()
partEmp = PartTimeEmployee("part time", "e3", 10, 100)
partEmp.calculateSalary()
```

#TASK 3
Design a Student Result System using encapsulation.
Requirements:
1. Private attribute __marks
2. Methods: set_marks(), get_marks(), calculate_grade()
3. Create at least two student objects

```python
class ResultSystem:
  def __init__(self, name):
    self.__marks = 0
    self.__name = name

  def setMarks(self, marks):
    self.__marks = marks

  def getMarks(self):
    return self.__marks

  def calculateGrade(self):
    if self.__marks >= 90:
      print(f"{self.__name}: grade A")
    elif self.__marks >= 80:
      print(f"{self.__name}: grade B")
    elif self.__marks >=70:
      print(f"{self.__name}: grade C")
    else:
      print(f"student {self.__name}: fail ")
student1 = ResultSystem("student1")
student1.setMarks(95)
student1.calculateGrade()
student2 = ResultSystem("student2")
student2.setMarks(50)
student2.calculateGrade()
```
